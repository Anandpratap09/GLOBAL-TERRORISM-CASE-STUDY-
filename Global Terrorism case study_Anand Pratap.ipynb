## Importing the Libraries:

#### Data Processing

```python

Global_terrorism.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'state',
                     'region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed',
                     'nwound': 'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                     'weaptype1_txt':'Weapon_type','motive':'Motive'
                    },inplace=True)
```

# As we have many columns, we take the columns that are necessary for analysis


```python
Global_terrorism = Global_terrorism[['eventid','Year','Month','Day','Country','Region','state','city','latitude','longitude','AttackType','Killed','Wounded',
            'Target','Summary','Group','Target_type','Weapon_type','Motive','success']]
```

```python
Global_terrorism['Killed'].sample(10)
```

#### Creating  a new column 'casualties' by adding 'killed' and 'wounded'

```python
Global_terrorism['casualities']=Global_terrorism['Killed']+Global_terrorism['Wounded']
Global_terrorism.head(3)
```

#### EDA(EXPLORATORY DATA ANALYSIS):

```python
year_attacks = Global_terrorism.groupby('Year').size().reset_index(name='count')
sns.lineplot(x='Year', y='count', data=year_attacks, color='red')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.title("Number of Terrorist Acticity")
plt.show()
```

```python
plt.subplots(figsize=(15,6))
sns.countplot(data=Global_terrorism, x='Year', palette='inferno')
plt.xticks(rotation=90)
plt.title('Number Of Terrorist Activities Each Year')
plt.show()
```

#### There has been a steady increase in global terrorist activities year on year. However, the year 2014 stands out as the peak with the highest recorded incidents. Post-2014, terrorist activity declined, which offers hope for improved global security.

#### TRENDS OF ATTACK AS PER REGION:

```python
year_attacks_region = Global_terrorism.groupby(['Year','Region']).size().reset_index(name='count')
```

```python
plt.subplots(figsize=(15,6))
sns.lineplot(x='Year',y='count',hue='Region',data=year_attacks_region)
plt.title('Terrorist Attacks Trends in Regions')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
plt.subplots(figsize=(15,6))
sns.countplot(x='Region',data=Global_terrorism,palette='magma',order=Global_terrorism['Region'].value_counts().index)
plt.xticks(rotation=80)
plt.title('Number Of Terrorist Activities By Region')
plt.show()
```

#### Terrorism in the Middle East has experienced repeated increases year after year, largely due to ongoing geopolitical conflicts and the presence of extremist groups. South Asia has also witnessed a rise in terrorism, often linked to criminal organizations and drug trafficking. Contrary to this, Central Asia has a relatively low rate of terrorism

#### Top 5 affected country:

```python
plt.subplots(figsize=(12,6))
top=Global_terrorism['Country'].value_counts()[:5].to_frame().reset_index()
top.columns= ['Country','Attacks_Counts']
sns.barplot(x='Country',y='Attacks_Counts', data= top, palette='magma')
plt.title('Top Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation=80)
plt.show()
```

#### Mostly affected Cities:

```python
plt.subplots(figsize=(12,6))
top=Global_terrorism['city'].value_counts()[:20].to_frame().reset_index()
top.columns= ['city','Attacks_Counts']
sns.barplot(x='city',y='Attacks_Counts', data= top, palette='magma')
plt.title('Top city Affected')
plt.xlabel('city')
plt.ylabel('Count')
plt.xticks(rotation=80)
```

#### Mode of Attack by the Terrorists: 

```python
plt.subplots(figsize=(15,6))
sns.countplot(x='AttackType',data=Global_terrorism,palette='inferno',order=Global_terrorism['AttackType'].value_counts().index)
plt.xticks(rotation=80)
plt.title('Attacking Methods by Terrorists')
plt.show()
```

#### Top Terror Groups:

```python
group_counts = Global_terrorism['Group'].value_counts()
sort = group_counts.sort_values(ascending=False)

# Select the top 5 most frequent groups
sort = sort.iloc[1:]
top_5 = sort.head(10)
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.barplot(x=top_5.index, y=top_5.values, palette="magma")
plt.title('Top Terrorist Attack Groups')
plt.xlabel('Attack Count')
plt.ylabel('Group')
plt.xticks(rotation = 80)
plt.show()
```

#### Mode of Avtivation of the Terrorist groups by yearon:

```python
top_groups10=Global_terrorism[Global_terrorism['Group'].isin(Global_terrorism['Group'].value_counts()[1:11].index)]
pd.crosstab(top_groups10.Year,top_groups10.Group).plot(color=sns.color_palette('Paired',10))
fig=plt.gcf()
fig.set_size_inches(16,6)
plt.show()
```

#### Casuality on yearly Basis:

```python
plt.subplots(figsize=(15,6))
year_cas = Global_terrorism.groupby('Year').casualities.sum().to_frame().reset_index()
year_cas.columns = ['Year','casualities']
sns.barplot(x=year_cas.Year, y=year_cas.casualities, palette='magma',edgecolor=sns.color_palette('dark',10))
plt.xticks(rotation=90)
plt.title('Number Of Casualities Each Year')
plt.show()
```

## Terrorist Attacks in India:

```python
India = Global_terrorism[(Global_terrorism['Country'] == 'India')]
India.head(5)
```

```python
India_attacks = India['eventid'].count()
print('There were',India_attacks ,'attacks in India.')
```

```python
India_success = India.groupby('success').size().reset_index(name='count')
India_success['percentage'] = India_success['count'] / India_attacks * 100
India_success
```

```python
sns.barplot(x='success', y='percentage', data = India_success,palette=['green', 'red'])
plt.title("Outcome of Terrorist Attacks in India")
plt.xlabel("Outcome")
```

## Mode of attacks and thier Sucess Ratio:

```python
attack_types_India = India.groupby(['AttackType','success']).size().reset_index(name='count')
attack_types_India
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x='AttackType', y='count', hue='success', data=attack_types_India,  palette= 'magma')
plt.xticks(rotation=30)
plt.title("Facility ")
```
